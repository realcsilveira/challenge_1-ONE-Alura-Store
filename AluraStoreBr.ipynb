### Importa√ß√£o dos dados



```python
import pandas as pd

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/main/base-de-dados-challenge-1/loja_1.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/main/base-de-dados-challenge-1/loja_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/main/base-de-dados-challenge-1/loja_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/main/base-de-dados-challenge-1/loja_4.csv"

loja1 = pd.read_csv(url)
loja2 = pd.read_csv(url2)
loja3 = pd.read_csv(url3)
loja4 = pd.read_csv(url4)

loja.head()
```

# 2. Vendas por Categoria


```python
print("Quantidade por categoria - Loja 1:")
print(loja1['Categoria do Produto'].value_counts())

print("\nQuantidade por categoria - Loja 2:")
print(loja2['Categoria do Produto'].value_counts())

print("\nQuantidade por categoria - Loja 3:")
print(loja3['Categoria do Produto'].value_counts())

print("\nQuantidade por categoria - Loja 4:")
print(loja4['Categoria do Produto'].value_counts())
```

# 3. M√©dia de Avalia√ß√£o das Lojas

```python
print("M√©dia das avalia√ß√µes:")
print("Loja 1:", loja1['Avalia√ß√£o da compra'].mean())
print("Loja 2:", loja2['Avalia√ß√£o da compra'].mean())
print("Loja 3:", loja3['Avalia√ß√£o da compra'].mean())
print("Loja 4:", loja4['Avalia√ß√£o da compra'].mean())
```

# 4. Produtos Mais e Menos Vendidos

```python
def mais_menos_vendidos(df, nome_loja):
    contagem = df['Produto'].value_counts()
    print(f"{nome_loja} - Mais vendido:\n{contagem.head(1)}")
    print(f"{nome_loja} - Menos vendido:\n{contagem.tail(1)}\n")

mais_menos_vendidos(loja1, "Loja 1")
mais_menos_vendidos(loja2, "Loja 2")
mais_menos_vendidos(loja3, "Loja 3")
mais_menos_vendidos(loja4, "Loja 4")
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

üìä 1. Gr√°fico de barras ‚Äì Faturamento total por loja

Este gr√°fico mostra o faturamento total de cada loja, permitindo uma compara√ß√£o clara de desempenho financeiro:

```python
# Faturamento total por loja
faturamento = {
    "Loja 1": loja1["Pre√ßo"].sum(),
    "Loja 2": loja2["Pre√ßo"].sum(),
    "Loja 3": loja3["Pre√ßo"].sum(),
    "Loja 4": loja4["Pre√ßo"].sum()
}

plt.figure(figsize=(8, 5))
sns.barplot(x=list(faturamento.keys()), y=list(faturamento.values()))
plt.title("Faturamento Total por Loja")
plt.ylabel("Faturamento (R$)")
plt.xlabel("Loja")
plt.tight_layout()
plt.savefig("imagens/faturamento_total.png", dpi=300, bbox_inches='tight')
plt.show()
```

üìà 2. Gr√°fico de linhas ‚Äì M√©dia das avalia√ß√µes dos clientes

Esse gr√°fico destaca o n√≠vel de satisfa√ß√£o dos clientes com base na m√©dia das avalia√ß√µes:

```python
# Avalia√ß√µes m√©dias por loja
avaliacoes = {
    "Loja 1": loja1["Avalia√ß√£o da compra"].mean(),
    "Loja 2": loja2["Avalia√ß√£o da compra"].mean(),
    "Loja 3": loja3["Avalia√ß√£o da compra"].mean(),
    "Loja 4": loja4["Avalia√ß√£o da compra"].mean()
}

plt.figure(figsize=(8, 5))
plt.plot(list(avaliacoes.keys()), list(avaliacoes.values()), marker='o', color='teal')
plt.title("M√©dia das Avalia√ß√µes por Loja")
plt.ylabel("Avalia√ß√£o M√©dia")
plt.xlabel("Loja")
plt.ylim(0, 5)
plt.grid(True)
plt.tight_layout()
plt.savefig("imagens/avaliacoes_media_lojas.png", dpi=300, bbox_inches='tight')
plt.show()
```

üìâ 3. Gr√°fico de barras horizontais ‚Äì Categorias mais vendidas (Loja 1)

üì¶ 4. Boxplot ‚Äì Distribui√ß√£o do custo de frete por loja

Esse gr√°fico mostra a distribui√ß√£o dos custos de frete, incluindo poss√≠veis outliers:

```python
# Concatenar os dados de todas as lojas com r√≥tulo
loja1['Loja'] = 'Loja 1'
loja2['Loja'] = 'Loja 2'
loja3['Loja'] = 'Loja 3'
loja4['Loja'] = 'Loja 4'

todas_lojas = pd.concat([loja1, loja2, loja3, loja4])

plt.figure(figsize=(10, 6))
sns.boxplot(data=todas_lojas, x='Loja', y='Frete')
plt.title("Distribui√ß√£o do Custo de Frete por Loja")
plt.ylabel("Custo de Frete (R$)")
plt.grid(axis='y')
plt.tight_layout()
plt.savefig("imagens/custo-frete.png", dpi=300, bbox_inches='tight')
plt.show()

```

üíµ 5. Gr√°fico de dispers√£o ‚Äì Rela√ß√£o entre pre√ßo e avalia√ß√£o dos produtos

Esse gr√°fico ajuda a entender se produtos mais caros tendem a ter avalia√ß√µes melhores (ou piores):

üßÆ 6. Gr√°fico de barras empilhadas ‚Äì Comparativo de categorias por loja
