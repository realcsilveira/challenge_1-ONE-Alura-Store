### Importação dos dados



```python
import pandas as pd

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/main/base-de-dados-challenge-1/loja_1.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/main/base-de-dados-challenge-1/loja_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/main/base-de-dados-challenge-1/loja_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/main/base-de-dados-challenge-1/loja_4.csv"

loja1 = pd.read_csv(url)
loja2 = pd.read_csv(url2)
loja3 = pd.read_csv(url3)
loja4 = pd.read_csv(url4)

loja.head()
```

```python
print("Faturamento total por loja:")
print("Loja 1:", loja1['Preço'].sum())
print("Loja 2:", loja2['Preço'].sum())
print("Loja 3:", loja3['Preço'].sum())
print("Loja 4:", loja4['Preço'].sum())
```

# 2. Vendas por Categoria


```python
print("Quantidade por categoria - Loja 1:")
print(loja1['Categoria do Produto'].value_counts())

print("\nQuantidade por categoria - Loja 2:")
print(loja2['Categoria do Produto'].value_counts())

print("\nQuantidade por categoria - Loja 3:")
print(loja3['Categoria do Produto'].value_counts())

print("\nQuantidade por categoria - Loja 4:")
print(loja4['Categoria do Produto'].value_counts())
```

# 3. Média de Avaliação das Lojas

```python
print("Média das avaliações:")
print("Loja 1:", loja1['Avaliação da compra'].mean())
print("Loja 2:", loja2['Avaliação da compra'].mean())
print("Loja 3:", loja3['Avaliação da compra'].mean())
print("Loja 4:", loja4['Avaliação da compra'].mean())
```

# 4. Produtos Mais e Menos Vendidos

```python
def mais_menos_vendidos(df, nome_loja):
    contagem = df['Produto'].value_counts()
    print(f"{nome_loja} - Mais vendido:\n{contagem.head(1)}")
    print(f"{nome_loja} - Menos vendido:\n{contagem.tail(1)}\n")

mais_menos_vendidos(loja1, "Loja 1")
mais_menos_vendidos(loja2, "Loja 2")
mais_menos_vendidos(loja3, "Loja 3")
mais_menos_vendidos(loja4, "Loja 4")
```

# 5. Frete Médio por Loja


```python
print("Custo médio de frete:")
print("Loja 1:", loja1['Frete'].mean())
print("Loja 2:", loja2['Frete'].mean())
print("Loja 3:", loja3['Frete'].mean())
print("Loja 4:", loja4['Frete'].mean())

```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

📊 1. Gráfico de barras – Faturamento total por loja

Este gráfico mostra o faturamento total de cada loja, permitindo uma comparação clara de desempenho financeiro:

```python
# Faturamento total por loja
faturamento = {
    "Loja 1": loja1["Preço"].sum(),
    "Loja 2": loja2["Preço"].sum(),
    "Loja 3": loja3["Preço"].sum(),
    "Loja 4": loja4["Preço"].sum()
}

plt.figure(figsize=(8, 5))
sns.barplot(x=list(faturamento.keys()), y=list(faturamento.values()))
plt.title("Faturamento Total por Loja")
plt.ylabel("Faturamento (R$)")
plt.xlabel("Loja")
plt.tight_layout()
plt.savefig("imagens/faturamento_total.png", dpi=300, bbox_inches='tight')
plt.show()
```

📈 2. Gráfico de linhas – Média das avaliações dos clientes

Esse gráfico destaca o nível de satisfação dos clientes com base na média das avaliações:

```python
# Avaliações médias por loja
avaliacoes = {
    "Loja 1": loja1["Avaliação da compra"].mean(),
    "Loja 2": loja2["Avaliação da compra"].mean(),
    "Loja 3": loja3["Avaliação da compra"].mean(),
    "Loja 4": loja4["Avaliação da compra"].mean()
}

plt.figure(figsize=(8, 5))
plt.plot(list(avaliacoes.keys()), list(avaliacoes.values()), marker='o', color='teal')
plt.title("Média das Avaliações por Loja")
plt.ylabel("Avaliação Média")
plt.xlabel("Loja")
plt.ylim(0, 5)
plt.grid(True)
plt.tight_layout()
plt.savefig("imagens/avaliacoes_media_lojas.png", dpi=300, bbox_inches='tight')
plt.show()
```

📦 4. Boxplot – Distribuição do custo de frete por loja

Esse gráfico mostra a distribuição dos custos de frete, incluindo possíveis outliers:

```python
# Concatenar os dados de todas as lojas com rótulo
loja1['Loja'] = 'Loja 1'
loja2['Loja'] = 'Loja 2'
loja3['Loja'] = 'Loja 3'
loja4['Loja'] = 'Loja 4'

todas_lojas = pd.concat([loja1, loja2, loja3, loja4])

plt.figure(figsize=(10, 6))
sns.boxplot(data=todas_lojas, x='Loja', y='Frete')
plt.title("Distribuição do Custo de Frete por Loja")
plt.ylabel("Custo de Frete (R$)")
plt.grid(axis='y')
plt.tight_layout()
plt.savefig("imagens/custo-frete.png", dpi=300, bbox_inches='tight')
plt.show()

```

💵 5. Gráfico de dispersão – Relação entre preço e avaliação dos produtos

Esse gráfico ajuda a entender se produtos mais caros tendem a ter avaliações melhores (ou piores):

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(data=todas_lojas, x='Preço', y='Avaliação da compra', hue='Loja', palette="Set2", alpha=0.6)
plt.title("Dispersão entre Preço e Avaliação dos Produtos")
plt.xlabel("Preço (R$)")
plt.ylabel("Avaliação do Cliente")
plt.grid(True)
plt.tight_layout()
plt.savefig("imagens/preco_avaliacao.png", dpi=300, bbox_inches='tight')
plt.show()
```

```python
# Contagem de categorias por loja
categorias_lojas = todas_lojas.groupby(['Loja', 'Categoria do Produto']).size().unstack().fillna(0)

# Gráfico de barras empilhadas
categorias_lojas.plot(kind='bar', stacked=True, figsize=(10, 6), colormap='tab20')
plt.title("Distribuição de Categorias de Produtos por Loja")
plt.ylabel("Quantidade de Produtos Vendidos")
plt.xlabel("Loja")
plt.legend(title='Categoria', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.savefig("imagens/categoria_loja.png", dpi=300, bbox_inches='tight')
plt.show()
```

# 📊 Relatório Final – Análise de Desempenho das Lojas

## 📝 Introdução

O objetivo deste trabalho foi analisar o desempenho de quatro lojas distintas com base em um conjunto de dados contendo informações sobre vendas, categorias de produtos, avaliações de clientes e frete. A partir dessas análises, nosso propósito é orientar o Senhor João sobre **qual loja ele deve vender**, considerando fatores como faturamento, popularidade dos produtos, satisfação dos clientes e custos operacionais.


## 💰 1. Faturamento Total por Loja

O faturamento total é um indicativo claro da performance comercial de cada loja. A análise revelou que:

- **Loja 4** teve o **maior faturamento**, destacando-se como a mais lucrativa.
- **Loja 1** e **Loja 3** apresentaram faturamentos intermediários.
- **Loja 2** teve o menor faturamento entre todas.

🔹 *Gráfico de barras abaixo mostra o comparativo de faturamento.*


```python
faturamento = {
    "Loja 1": loja1["Preço"].sum(),
    "Loja 2": loja2["Preço"].sum(),
    "Loja 3": loja3["Preço"].sum(),
    "Loja 4": loja4["Preço"].sum()
}

plt.figure(figsize=(8, 5))
sns.barplot(x=list(faturamento.keys()), y=list(faturamento.values()))
plt.title("Faturamento Total por Loja")
plt.ylabel("Faturamento (R$)")
plt.xlabel("Loja")
plt.tight_layout()
plt.show()
```

## 🛍️ 2. Categorias de Produtos Mais Vendidas

Analisamos quais categorias tiveram mais vendas em cada loja. A **Loja 4** foi a mais diversificada, enquanto a **Loja 1** concentrou maior volume em menos categorias.

🔹 *Abaixo, o gráfico de barras empilhadas mostra a distribuição por loja.*


```python
categorias_lojas = todas_lojas.groupby(['Loja', 'Categoria do Produto']).size().unstack().fillna(0)

categorias_lojas.plot(kind='bar', stacked=True, figsize=(10, 6), colormap='tab20')
plt.title("Distribuição de Categorias de Produtos por Loja")
plt.ylabel("Quantidade de Produtos Vendidos")
plt.xlabel("Loja")
plt.legend(title='Categoria', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```

## ⭐ 3. Avaliações dos Clientes

A satisfação dos clientes foi avaliada com base na média das notas dadas por eles. A **Loja 3** obteve a melhor média, seguida por Loja 1 e Loja 4. A **Loja 2** ficou atrás nesse aspecto.

🔹 *Gráfico de linha abaixo mostra a comparação das avaliações médias.*


```python
avaliacoes = {
    "Loja 1": loja1["Avaliação da compra"].mean(),
    "Loja 2": loja2["Avaliação da compra"].mean(),
    "Loja 3": loja3["Avaliação da compra"].mean(),
    "Loja 4": loja4["Avaliação da compra"].mean()
}

plt.figure(figsize=(8, 5))
plt.plot(list(avaliacoes.keys()), list(avaliacoes.values()), marker='o', color='teal')
plt.title("Média das Avaliações por Loja")
plt.ylabel("Avaliação Média")
plt.xlabel("Loja")
plt.ylim(0, 5)
plt.grid(True)
plt.tight_layout()
plt.show()
```

## 🚚 4. Custo Médio de Frete

A logística também impacta o negócio. A **Loja 3** teve o menor custo médio de frete. A **Loja 4**, embora com bom desempenho, teve o frete mais elevado.

🔹 *Boxplot abaixo representa a variação do frete por loja.*


```python
plt.figure(figsize=(10, 6))
sns.boxplot(data=todas_lojas, x='Loja', y='Frete')
plt.title("Distribuição do Custo de Frete por Loja")
plt.ylabel("Custo de Frete (R$)")
plt.grid(axis='y')
plt.tight_layout()
plt.show()
```

## 📉 5. Relação entre Preço e Avaliação

Verificamos se produtos mais caros têm avaliações melhores. A análise mostrou que **não há uma correlação clara** — produtos bem avaliados existem em todas as faixas de preço.

🔹 *Gráfico de dispersão abaixo mostra essa relação por loja.*


```python
plt.figure(figsize=(10, 6))
sns.scatterplot(data=todas_lojas, x='Preço', y='Avaliação da compra', hue='Loja', palette="Set2", alpha=0.6)
plt.title("Dispersão entre Preço e Avaliação dos Produtos")
plt.xlabel("Preço (R$)")
plt.ylabel("Avaliação do Cliente")
plt.grid(True)
plt.tight_layout()
plt.show()
```

Após analisar os principais indicadores de desempenho, a **recomendação é que o Senhor João venda a Loja 4**.
